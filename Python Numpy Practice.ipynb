```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
iris_2d = np.genfromtxt(url, dtype= "object", delimiter= ",")
# ["Sepallength" "sepelwidth" "petallength" "petalwidth" "species"]
sepallength = iris_2d[:, 0].astype('float')
petallength = iris_2d[:, 2].astype('float')
space = (np.pi * petallength * (sepallength**3))/4
space = space[:, np.newaxis]
up_iris_2d = np.hstack([iris_2d, space])
up_iris_2d
```
